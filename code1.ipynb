# 日期数据处理，格式转化，取年份

```python
air_data["year1"] = air_data["year"].map(lambda x: int(x/100+1))
```

```python
#计算哪一年积分最高
fig = plt.figure(figsize=(15, 5))
ax = fig.gca()
air_data.groupby("year")[["Points_Sum"]].mean().sort_values('Points_Sum', ascending=False).tail(20).plot.bar(stacked=False, ax=ax, color='darkorange')
#.mean().sort_values('Points_Sum', ascending=False)
```
